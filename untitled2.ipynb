```python
from transformers import pipeline

# Load a text2text generation pipeline with a pre-trained model
generator = pipeline("text2text-generation", model="google/flan-t5-large")

# Sample input - product review text
product_review = """
I recently bought the ABC Bluetooth speaker, and the sound quality is fantastic. The battery life lasts more than 12 hours on a single charge. However, the Bluetooth connection drops occasionally, and it takes time to reconnect.
"""

# Prompt engineering to guide the model
prompt = f"Extract pros and cons from the following product review:\n\n{product_review}"

# Generate the output
result = generator(prompt, max_length=150, clean_up_tokenization_spaces=True)[0]['generated_text']

print("=== Pros and Cons Extracted ===")
print(result)
```
